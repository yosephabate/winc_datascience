# Matplotlib Practice

```python
# The default graph of matplotlib is "line plot"
```

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# subplots() returns a tuple containing two objects
#  fig: representing the whole graph
#  ax: representing each sub-graph axes

x_values = [1, 2, 3, 4]
y_values = [1, 4, 2, 3]

fig, ax = plt.subplots()
ax.plot(x_values, y_values)
```

```python
# old way of plotting
# not recommended
plt.plot(x_values, y_values)
```

```python
# multiple plots in a figure
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
x = np.arange(0, 4*math.pi, step=0.1)
ax1.plot(x, x * x)
ax2.plot(x, np.sin(x))
ax3.plot(x, 1/x)
ax4.plot(x, np.cos(x))
```

```python
fig, ax = plt.subplots()
names = ['Abebe', 'Kebede', 'Halima', 'Yoseph']
ages = [24, 30, 20, 39]
ax.plot(names, ages)
```

```python
# multiple lines in a plot
fig, ax = plt.subplots()
x = np.arange(0, 2*math.pi, step=0.1)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

```python
# labeling lines
fig, ax = plt.subplots()
x = np.arange(0, 2*math.pi, step=0.1)
ax.plot(x, np.sin(x), label='sin(x)')
ax.plot(x, np.cos(x), label='cos(x)')
ax.legend()
```

```python
# set x and y limits
fix, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
x = np.arange(0, 4*math.pi, step=0.1)
ax1.plot(x, x * x)
ax2.plot(x, np.sin(x))
ax3.plot(x, x**2 * 1.5)
ax4.plot(x, np.cos(x))

ax1.set_ylim(-1, 30)
ax2.set_ylim(-1, 30)
ax3.set_ylim(-1, 30)
ax4.set_ylim(-1, 30)
```

```python
# labeling the x and y axes
fig, ax = plt.subplots()
x = np.arange(1, 13)
ax.plot(x, x + 3, label="Wedding Ceremony")
ax.plot(x, np.sin(x) + x, label="Sunday Worship")
ax.legend()

ax.set_xlim(1, 12)
ax.set_ylim(0, 20)

ax.set_xlabel('month')
ax.set_ylabel('attendance')
```

```python
# formatting markers and lines
fig, ax = plt.subplots()
x = np.arange(1, 13)
ax.plot(x, x + 3, label="Wedding Ceremony", color='magenta', marker='.', linestyle='--')
ax.plot(x, np.sin(x) + x, label="Sunday Worship", color='green', marker='8', linestyle='')
ax.legend()

ax.set_xlim(1, 12)
ax.set_ylim(0, 20)

ax.set_xlabel('month')
ax.set_ylabel('attendance')
```

```python
# formatting markers and lines - short version
# right after the x and y values (plot function)
fig, ax = plt.subplots()
x = np.arange(1, 13)
ax.plot(x, x + 3, '.--m', label="Wedding Ceremony")
ax.plot(x, np.sin(x) + x, '8g', label="Sunday Worship")
ax.legend()

ax.set_xlim(1, 12)
ax.set_ylim(0, 20)

ax.set_xlabel('month')
ax.set_ylabel('attendance')
```

```python
# set ax a title
fig, ax = plt.subplots()
x = np.arange(1, 13)
ax.plot(x, x + 3, '.--m', label="Wedding Ceremony")
ax.plot(x, np.sin(x) + x, '8g', label="Sunday Worship")
ax.legend()

ax.set_xlim(1, 12)
ax.set_ylim(0, 20)

ax.set_xlabel('month')
ax.set_ylabel('attendance')

ax.set_title('Church Goers Attendance After COVID-19')
```

```python
# show grid
fig, ax = plt.subplots()
x = np.arange(1, 13)
ax.plot(x, x + 3, '.--m', label="Wedding Ceremony")
ax.plot(x, np.sin(x) + x, '8g', label="Sunday Worship")
ax.legend()

ax.set_xlim(1, 12)
ax.set_ylim(0, 20)

ax.set_xlabel('month')
ax.set_ylabel('attendance')

ax.set_title('Church Goers Attendance After COVID-19')
ax.grid(True)
```
