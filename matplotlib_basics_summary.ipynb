```python
# important libraries for graph plotting, pandas and numpy operations
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math

x = np.arange(0, 6*math.pi, 0.1)

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2)
fig.set_size_inches(15, 15)

ax1.plot(x, x**2, label='x**2', color='magenta', linestyle='--', marker='.')
ax1.legend()
ax1.set_xlabel('x')
ax1.set_ylabel('x**2')
ax1.set_title('The graphical representation of x**2')

ax2.plot(x, 1/x, label='1/x', color='chocolate')
ax2.legend()

ax3.plot(x, np.log10(x), label='log10(x)', color='lightseagreen')
ax3.legend()

ax4.plot(x, np.tan(x), label='tan(x)', color='dodgerblue')
ax4.legend()

ax5.plot(x, np.sin(x), label='sin(x)', color='limegreen')
ax5.legend()

ax6.plot(x, np.cos(x), label='cos(x)', color='darkorange')
ax6.legend()

```
