```python
import numpy as np
import pandas as pd
```

```python
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_by_birth_rate")
countries = tables[0]
countries
```

```python
countries.columns = ['country', 'wb2018', 'oecd2011', 'ciawf2013', 'ciawf2014', 'ciawf2020', 'prb2020']
```

```python
countries_name = countries.set_index('country')
```

```python
countries_name.loc[['Ethiopia','Eritrea'], :]
```

```python
birth_rate_more_than_30_percent = countries_name.prb2020 >= 40
countries_name.loc[birth_rate_more_than_30_percent]
```

```python
def name_longer_than_20(name):
  return len(name) > 20

countries.loc[countries.country.apply(name_longer_than_20)]
```

```python
countries.loc[~countries.country.str.contains('Anguilla')]
```

```python
prb_more_than_30 = countries.prb2020 > 30
countries.loc[prb_more_than_30, ['country', 'prb2020']]
```

```python
countries_with_stan = countries.country.str.contains('stan')
countries_lower_than_25 = countries.ciawf2020 < 25
countries.loc[countries_with_stan & countries_lower_than_25, ['country', 'ciawf2020']]
```

```python
def dropped_more_than_7_percent(country):
  drop_value = country.ciawf2013 - country.ciawf2020
  return drop_value > 7

countries.loc[lambda c: dropped_more_than_7_percent(c), ['country', 'ciawf2013', 'ciawf2020']]
```
