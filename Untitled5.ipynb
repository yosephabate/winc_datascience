```python
def is_name_more_than_12_letters(name):
  return len(name) > 12
```

```python
is_name_more_than_12_letters('Abebe')
```

```python
is_name_more_than_12_letters('Abebe is coming home today')
```

```python

smoking.loc[smoking.borough.apply(is_name_more_than_12_letters)]
```

```python
names_longer_than_12 = smoking.borough.apply(is_name_more_than_12_letters)
smoking.loc[~names_longer_than_12]
```

```python
def more_than_23_percent_smokers(sample):
  percent = (sample.current / sample.total) * 100
  return percent > 23
```

```python
smoking.loc[lambda row: more_than_23_percent_smokers(row)]
```

```python
total_more_than_200_000 = smoking.total > 200_000
smoking.loc[total_more_than_200_000]
```

```python
my_condition = (smoking.current > 50_000) | (smoking.total > 220_000)
smoking.loc[my_condition]
```

```python
my_condition = (smoking.current > 50_000) & (smoking.total > 220_000)
smoking.loc[my_condition]
```

```python
within_these_cities = smoking.borough.isin(['Croydon', 'Addis Ababa', 'Hackney'])
smoking.loc[within_these_cities]
```

```python
containing_string_ham = smoking.borough.str.contains('ham')
smoking.loc[containing_string_ham]
```
